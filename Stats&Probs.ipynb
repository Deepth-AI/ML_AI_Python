#Day3- Probability & Statistics

```python
#Day3- Probability & Statistics

import numpy as np
import pandas as pd
import seaborn as sns

df=pd.read_csv("/content/drive/MyDrive/Datasets/titanic.csv")

```

#Basic Statistics

```python
print(df.describe())
print(df['Age'].mean())
print(df['Age'].std())
print(df['Age'].var())
print(df['Age'].min())
print(df['Age'].max())
print(df['Age'].median())
```

#Probability example â€“ survival rate


```python
survival_prob = df['Survived'].mean()
survival_prob
print(f"survival_rate:{survival_prob:.2%}")
```

#Distribution plot of Age


```python
sns.histplot(df['Age'].dropna(), kde=True)
```


#Compare survival by Sex

```python
sns.barplot(x='Sex', y='Survived', data=df)
```
